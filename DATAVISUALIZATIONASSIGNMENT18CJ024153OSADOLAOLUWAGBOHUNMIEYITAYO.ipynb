# Number of Combined Radio and Television Institutions in Africa

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python

df   = pd.read_csv('rttt.csv')
df = df.set_index('Reference Area')
# Total Number of Combined Radio and Television Institutions in Africa

df
```

```python
for year in range(2010,2020):
    df.plot.barh(y=str(year), figsize=(10,10), title=str(year))
    plt.figure(figsize = (40,40))
```

```python
cols = ['2010','2011','2012','2013','2014','2015','2016','2017','2018','2019']
```

```python
df[cols].mean()
```

```python
plt.bar(cols,df[cols].mean())
plt.xlabel('Year')
plt.ylabel('Mean')
plt.title('Mean of the Combined Radio and Television Institutions')
plt.figure(figsize = (20,20))
```

```python
plt.plot(cols,df[cols].mean())
plt.xlabel('Year')
plt.ylabel('Mean')
plt.title('Mean of the Combined Radio and Television Institutions')
```

```python
df[cols].std()
# to find the standard deviation of each year
```

```python
plt.bar(cols,df[cols].std())

plt.xlabel('Year')
plt.ylabel('Standard Deviation')
plt.title('Standard Deviation of Combined Radio and Television Institutions')
```

```python
plt.plot(cols,df[cols].std())
plt.xlabel('Year')
plt.ylabel('Standard Deviation')
plt.title('Standard Deviation of Combined Radio and Television Institutions')
plt.figure(figsize = (10,10))
```

```python
df[cols].var()
```

```python
plt.plot(cols,df[cols].var())
plt.title('Variance of the Comnined Radio and Public Institutions')
plt.xlabel('Year')
plt.ylabel('Variance')
plt.figure(figsize = (10,10))
```

```python
plt.bar(cols,df[cols].var())
plt.xlabel('Year')
plt.ylabel('Variance')
plt.title('Variance of the Comnined Radio and Public Institutions')
plt.figure(figsize = (10,10))
```

```python
df[cols].median()
#median of each year of combined radio and television institutions
```

```python
plt.bar(cols,df[cols].median())
plt.xlabel('Year')
plt.ylabel('Median')
plt.title('Median of combined radio and television institutions')
plt.figure(figsize = (20,20))
```

```python
plt.plot(cols,df[cols].median())
plt.xlabel('Year')
plt.ylabel('Median')
plt.title('Median of combined radio and television institutions')
plt.figure(figsize = (10,10))
```

```python
iqrs = []
for year in range(2010,2020):
    q3,q1 = np.percentile(df[str(year)], [75,25])
    iqrs.append(q3-q1)

print(cols)
print(iqrs)
plt.plot(cols, iqrs)
plt.ylabel('InterquartileRange')
plt.xlabel('Years')
plt.title('Interquartile Range of the combined radio and television institutions in Africa')
```
